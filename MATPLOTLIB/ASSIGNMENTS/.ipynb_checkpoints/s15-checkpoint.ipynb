```python
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
df = pd.read_csv("spotify.csv")
profile = ProfileReport(df, title="Spotify Top 100 Songs Report")
profile.to_file("spotify_report.html")
```

```python
pip install sweetviz
```

```python
import sweetviz as sv

mumbai = pd.read_csv("mumbai.csv")
delhi = pd.read_csv("delhi.csv")
report = sv.compare(
    [mumbai, "Mumbai"],
    [delhi, "Delhi"]
)
report.show_html("comparison_report.html")
```
