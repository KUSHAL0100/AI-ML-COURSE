```python
"""
1.
Use plt.subplots() to create a 2x2 grid of subplots and plot four different types of charts (line, bar, scatter, and pie) using any sample data of your choice.
2.
Plot a comparison of average delivery times for Zomato, Swiggy, and Domino's using a bar chart, and style each bar with a different color, linestyle, and linewidth using Matplotlib plot styling options.
3.
Create a multi-axis chart that shows the number of Instagram followers (left y-axis) and average daily posts (right y-axis) for five influencers. Use different colors and markers for each axis.
4.
Plot the number of tickets sold for five recent Bollywood movies (categorical) versus their IMDB ratings (numeric) using a scatter plot. Add annotations to display the movie names above each point.
5.
Add a custom text annotation to a Matplotlib chart showing Flipkart's monthly sales, marking the highest sales point with the label 'Big Billion Days'.<br><br><em><strong>Hint:</strong> Use the ax.annotate() function to place the label at the correct data point.</em>
"""
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.figure(figsize=(8,6))

# Line Plot
plt.subplot(2,2,1)
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Bar Chart
plt.subplot(2,2,2)
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Scatter Plot
plt.subplot(2,2,3)
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Pie Chart
students = [25,30,20,25]
courses = ["Python","Java","C++","Web"]

plt.subplot(2,2,4)
plt.pie(students,
        labels=courses,
        autopct="%1.1f%%")
plt.title("Students in Courses")

plt.tight_layout()
plt.show()
```

```python
apps = ["Zomato","Swiggy","Domino's"]
time = [35,30,40]

plt.bar(apps,time,
        color=["red","orange","blue"],
        edgecolor="black",
        linewidth=2)

plt.title("Average Delivery Time")
plt.xlabel("Delivery Apps")
plt.ylabel("Time (Minutes)")

plt.show()
```

```python
name = ["A","B","C","D","E"]
followers = [100,150,200,180,250]
posts = [10,15,5,25,19]

plt.plot(name, followers, marker="o", color="blue", label="Followers")
plt.plot(name, posts, marker="*", color="red", label="Daily Posts")

plt.title("Followers vs Daily Posts")
plt.xlabel("Influencers")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
movie = ["Sitaare","Raid 2","Housefull","Maa","Saiyaara"]
tickets = [100,150,180,200,160]
rating = [7.5,8.2,7.8,9.0,8.0]

plt.scatter(tickets,rating)

for i in range(5):
    plt.text(tickets[i],rating[i],movie[i])

plt.title("Bollywood Movies")
plt.xlabel("Tickets Sold (Lakhs)")
plt.ylabel("IMDB Rating")

plt.show()
```

```python
month = ["Jan","Feb","Mar","Apr","May"]
sales = [50,70,120,80,60]

plt.plot(month, sales, marker="o")

plt.title("Flipkart Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales (Crores)")

plt.annotate("Big Billion Days",
             xy=("Mar",120),
             xytext=("Apr",130),
             arrowprops=dict(arrowstyle="->"))

plt.grid(True)

plt.show()
```
