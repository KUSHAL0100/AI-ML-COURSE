```python
"""
1.
Install Seaborn in your Python environment and use it to create a histplot showing the distribution of delivery times (in minutes) for 50 Zomato food orders. Generate random data if needed.
2.
Load the 'tips' dataset from Seaborn and create a boxplot to visualize the distribution of total bill amounts by day of the week, similar to how Zomato might analyze spending patterns across weekdays.
3.
Simulate IPL match scores for 8 teams and use a violinplot to compare the run distributions per team. Style your plot using the 'darkgrid' Seaborn theme.
4.
Create a countplot that displays the number of songs per genre from a list of 40 Spotify tracks, with at least 4 different genres represented.<br><br><em><strong>Hint:</strong> Use a Python list or pandas DataFrame to store your data, then plot with Seaborn.</em>
5.
Use Seaborn's kdeplot to visualize the distribution of daily step counts for a week, as if analyzing data from a fitness app. Apply the 'whitegrid' theme and customize the plot color.
"""
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
delivery_time = np.random.randint(20, 60, 50)

sns.histplot(delivery_time, bins=10, color="blue")

plt.title("Zomato Delivery Time Distribution")
plt.xlabel("Delivery Time (Minutes)")
plt.ylabel("Frequency")

plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.set_theme(style="darkgrid")

teams = ["CSK","MI","RCB","GT","KKR","RR","DC","PBKS"]

data = {
    "Team": np.repeat(teams, 10),
    "Runs": np.random.randint(120, 230, 80)
}

df = pd.DataFrame(data)

sns.violinplot(x="Team", y="Runs", data=df)

plt.title("IPL Team Run Distribution")
plt.xlabel("Teams")
plt.ylabel("Runs")

plt.show()
```

```python
genre = [
"Pop","Rock","Hip-Hop","Classical",
"Pop","Rock","Pop","Hip-Hop",
"Classical","Pop","Rock","Hip-Hop",
"Pop","Rock","Classical","Pop",
"Hip-Hop","Rock","Pop","Classical",
"Rock","Hip-Hop","Pop","Rock",
"Classical","Pop","Hip-Hop","Rock",
"Pop","Rock","Classical","Hip-Hop",
"Pop","Rock","Pop","Classical",
"Hip-Hop","Rock","Pop","Pop"
]

df = pd.DataFrame({"Genre": genre})

sns.countplot(x="Genre", data=df)

plt.title("Songs by Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")

plt.show()
```

```python
sns.set_theme(style="whitegrid")

steps = [4500,5200,6100,7000,6500,8000,7500]

sns.kdeplot(steps, fill=True, color="green")

plt.title("Daily Step Count Distribution")
plt.xlabel("Steps")
plt.ylabel("Density")

plt.show()
```
