```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("IPL.csv")
print(df['runs_total'].head(20))
print("Mean =", df["runs_total"].mean())
print("Median =", df["runs_total"].median())
print("Minimum =", df["runs_total"].min())
print("Maximum =", df["runs_total"].max())
print("Standard Deviation =", df["runs_total"].std())
```

```python
df = pd.read_csv("flipkart_orders.csv")
sns.countplot(x="rating", data=df)

plt.title("Count of Reviews by Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")

plt.show()
```

```python
df=pd.read_csv('zomato.csv')
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="average_cost_for_two",
    y="user_rating",
    s=100
)

plt.title("Relationship Between Average Cost for Two and User Rating")
plt.xlabel("Average Cost for Two (₹)")
plt.ylabel("User Rating")
plt.grid(True)

plt.show()
```

```python
# Load dataset
df = pd.read_csv("bookmyshow.csv")

# Group by genre and calculate average box office collection
avg_collection = (
    df.groupby("genre")["box_office_collection"]
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)
print(avg_collection)
```

```python
# Load dataset
df = pd.read_csv("spotify.csv")

# Create pairplot
sns.pairplot(
    df[["danceability", "energy", "valence", "popularity"]],
    diag_kind="hist" #kind and diag_kind is different
)

plt.suptitle("Pairwise Relationships Among Spotify Song Features",y=1.02)
plt.show()
```
