```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.figure(figsize=(8,6))

# Line Plot
plt.subplot(2,2,1)
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Bar Chart
plt.subplot(2,2,2)
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Scatter Plot
plt.subplot(2,2,3)
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Pie Chart
students = [25,30,20,25]
courses = ["Python","Java","C++","Web"]

plt.subplot(2,2,4)
plt.pie(students,
        labels=courses,
        autopct="%1.1f%%")
plt.title("Students in Courses")

plt.tight_layout()
plt.show()
```

```python
name = ["A","B","C","D","E"]
followers = [100,150,200,180,250]
posts = [10,15,5,25,19]

plt.plot(name, followers, marker="o", color="blue", label="Followers")
plt.plot(name, posts, marker="*", color="red", label="Daily Posts")

plt.title("Followers vs Daily Posts")
plt.xlabel("Influencers")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
movie = ["Sitaare","Raid 2","Housefull","Maa","Saiyaara"]
tickets = [100,150,180,200,160]
rating = [7.5,8.2,7.8,9.0,8.0]

plt.scatter(tickets,rating)

for i in range(5):
    plt.text(tickets[i],rating[i],movie[i])

plt.title("Bollywood Movies")
plt.xlabel("Tickets Sold (Lakhs)")
plt.ylabel("IMDB Rating")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
delivery_time = np.random.randint(20, 60, 50)

sns.histplot(delivery_time, bins=10, color="blue")

plt.title("Zomato Delivery Time Distribution")
plt.xlabel("Delivery Time (Minutes)")
plt.ylabel("Frequency")

plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)

plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")

plt.show()
```

```python
sns.set_theme(style="darkgrid")

teams = ["CSK","MI","RCB","GT","KKR","RR","DC","PBKS"]

data = {
    "Team": np.repeat(teams, 10),
    "Runs": np.random.randint(120, 230, 80)
}

df = pd.DataFrame(data)

sns.violinplot(x="Team", y="Runs", data=df)

plt.title("IPL Team Run Distribution")
plt.xlabel("Teams")
plt.ylabel("Runs")

plt.show()
```

```python
genre = [
"Pop","Rock","Hip-Hop","Classical",
"Pop","Rock","Pop","Hip-Hop",
"Classical","Pop","Rock","Hip-Hop",
"Pop","Rock","Classical","Pop",
"Hip-Hop","Rock","Pop","Classical",
"Rock","Hip-Hop","Pop","Rock",
"Classical","Pop","Hip-Hop","Rock",
"Pop","Rock","Classical","Hip-Hop",
"Pop","Rock","Pop","Classical",
"Hip-Hop","Rock","Pop","Pop"
]

df = pd.DataFrame({"Genre": genre})

sns.countplot(x="Genre", data=df)

plt.title("Songs by Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")

plt.show()
```

```python
sns.set_theme(style="whitegrid")

steps = [4500,5200,6100,7000,6500,8000,7500]

sns.kdeplot(steps, fill=True, color="green")

plt.title("Daily Step Count Distribution")
plt.xlabel("Steps")
plt.ylabel("Density")

plt.show()
```
