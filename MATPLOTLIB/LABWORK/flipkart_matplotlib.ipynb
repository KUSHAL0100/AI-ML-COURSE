```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('flipkart_orders.csv')
```

```python
df.columns.tolist()
```

```python
print(df["category"].unique())
print(df["payment_method"].unique())
print(df["order_status"].unique())
print(df["state"].unique())
print(df["city"].unique())


# faaaaaaaa,category
```

```python
#CLEAN COLUMNS, BOOKS and books should be same naaaa
df["category"] = df["category"].str.strip().str.title()
```

```python
result=df.groupby("category")["price"].mean()
plt.figure(figsize=(15,8))
plt.plot(result.index,result.values)
plt.xlabel('Category')
plt.ylabel('Price')
plt.title('category vs avg price')
plt.show()
```

```python
result=df.groupby("category")["revenue"].sum()
plt.figure(figsize=(15,8))
plt.bar(result.index,result.values)
plt.xlabel('Category')
plt.ylabel('revenue')
plt.title('category vs revenue')
plt.show()
```

```python
# FREQUENCY OF PRICE
plt.figure(figsize=(10,6))

plt.hist(df["price"], bins=20)

plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")

plt.show()
```

```python
#revenue state wise
result=df.groupby('state')['revenue'].mean()
plt.figure(figsize=(15,10))
plt.bar(result.index,result.values)
plt.xlabel('state')
plt.ylabel('revenue')
plt.title('avg state wise revenue')
plt.show()
```

```python
# city wise total orders
result=df.groupby('city')['order_id'].count()
plt.figure(figsize=(15,10))
plt.bar(result.index,result.values)
plt.xlabel('city')
plt.ylabel('orders')
plt.title('City wise Orders')
plt.show()
```

```python
#Age Frequency
plt.figure(figsize=(15,8))
plt.hist(df['age'],edgecolor="black")
plt.xticks(range(int(df['age'].min()),int(df['age'].max()), 5)) #THIS IS BETTER THAN BINS,lmaooo
plt.xlabel('age')
plt.ylabel('frequecy')
plt.title('age frequency')
plt.show()
```
